```python
#SPEECH TO TEXT
! pip install git+https://github.com/huggingface/transformers -q
from transformers import pipeline
#Initializes the SpeechRecognizer class with a specified model for speech recognition.
class SpeechRecognizer:
    def __init__(self, model_name="openai/whisper-large-v3"):
        self.whisper_asr = pipeline("automatic-speech-recognition", model=model_name)
#Performs speech recognition on an audio file.
    def recognize_speech(self, audio_path):
        result = self.whisper_asr(audio_path)
        return result['text']

# Example usage
if __name__ == "__main__":
    recognizer = SpeechRecognizer()
    audio_path = '/content/input.mp3'
    transcription = recognizer.recognize_speech(audio_path)
    print(transcription)
```

```python

!pip install git+https://github.com/suno-ai/bark.git
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio
# Preloading the models for faster audio generation
preload_models()
```

```python
def generate_and_display_audio(text):
    """
    Generates audio from the given text prompt and displays it.
    """
    audio_array = generate_audio(text)
    return Audio(audio_array, rate=SAMPLE_RATE)

# Example usage
text_prompt = """
नमस्ते, मेरा नाम आरव है। मुझे समोसा बहुत पसंद है और मैं अक्सर इसे दोस्तों के साथ खाता हूँ। [हंसते हैं] लेकिन मुझे अन्य चीज़ों में भी रुचि है, जैसे कि क्रिकेट खेलना। जब भी मुझे समय मिलता है, मैं क्रिकेट खेलने के लिए मैदान पर जाता हूँ। क्रिकेट मेरे लिए सिर्फ एक खेल नहीं है, बल्कि एक जुनून है। मुझे लगता है कि यह मुझे शारीरिक और मानसिक रूप से फिट रखता है। साथ ही, टीम वर्क और रणनीति बनाने की कला भी सिखाता है।
"""
generate_and_display_audio(text_prompt)
```
